## ✅**Parte 4: Función para el procesamiento de información**

**```4```**

````python
def procesar_padron_ruc(mes, anio):

    """
    Procesa el padrón RUC de un mes y año específicos.

    Args:
        mes (str): El nombre del mes en minúsculas (e.g., 'julio').
        año (str): El año en formato de cuatro dígitos (e.g., '2024').

    Returns:
        pd.DataFrame: DataFrame con las empresas filtradas y procesadas.
    """

    # Diccionario de meses
    meses_dict = {
                'enero': '01', 'febrero': '02', 'marzo': '03', 'abril': '04',
                'mayo': '05', 'junio': '06', 'julio': '07', 'agosto': '08',
                'septiembre': '09', 'octubre': '10', 'noviembre': '11', 'diciembre': '12'
                 }

    # obtener el numero del mes y asegurar que está en minuscula
    mes_num = meses_dict.get(mes.lower())

    # Formar la ruta del archivo con los args
    ruta = f"../data/raw/PadronRUC_{anio}{mes_num}.csv"

    # Cargar el archivo CSV
    padron = pd.read_csv(ruta, encoding='latin1')

    # Crear una nueva columna llamada 'region'
    padron['region'] = ''

    # Asignar el departamento en región correspondiente
    padron.loc[padron['Departamento'].isin(
        ['LIMA', 'CALLAO', 'PIURA', 'LAMBAYEQUE', 'TUMBES', 'ICA', 'LA LIBERTAD']), 'region'] = 'Costa'
    padron.loc[padron['Departamento'].isin(
        ['CUSCO', 'HUANCAVELICA', 'HUANUCO', 'JUNIN', 'AYACUCHO', 'ANCASH', 'PUNO', 'CAJAMARCA', 'TACNA', 'APURIMAC', 'MOQUEGUA', 'PASCO' , 'AREQUIPA']), 'region'] = 'Sierra'
    padron.loc[padron['Departamento'].isin(
        ['LORETO', 'UCAYALI', 'MADRE DE DIOS', 'SAN MARTIN', 'AMAZONAS']), 'region'] = 'Selva'

    # Filtrar solo las empresas que pertenecen a la Costa
    padron_costa = padron[padron['region'] == 'Costa']

    # Crear la columna 'comercio_resumen' a partir de 'ComercioExterior'
    padron_costa['comercio_resumen'] = " "
    padron_costa.loc[padron_costa['ComercioExterior'].isin(['IMPORTADOR' , 'EXPORTADOR' , 'IMPORTADOR/EXPORTADOR' ]) , "comercio_resumen" ] = "Con actividad"
    padron_costa.loc[padron_costa['ComercioExterior'].isin(['SIN ACTIVIDAD' , 'NO DISPONIBLE' ]) , "comercio_resumen" ] = 'Sin actividad'

    # Reemplazar los valores "NO DISPONIBLE" por NaN en la columna 'NroTrab'
    padron_costa['NroTrab'] = padron_costa['NroTrab'].replace("NO DISPONIBLE", np.nan)

    # Eliminar las filas con valores NaN en 'NroTrab'
    padron_costa = padron_costa.dropna(subset=['NroTrab'])

    # Convertir la columna 'NroTrab' a tipo Int64
    padron_costa['NroTrab'] = padron_costa['NroTrab'].astype('Int64')

    # Filtrar empresas con menos de 1000 trabajadores (ignorando nulos)
    padron_costa = padron_costa[padron_costa['NroTrab'] < 1000]

    # Convertir el RUC a cadena de texto (str) y filtrar empresas cuyo RUC empieza con '2'
    padron_costa['RUC'] = padron_costa['RUC'].astype(str)
    padron_costa = padron_costa[padron_costa['RUC'].str.startswith('2')]

    # Seleccionar las columnas necesarias
    columnas = ['RUC', 'TipoFacturacion', 'TipoContabilidad', 'Departamento', 'region', 'NroTrab', 'comercio_resumen']
    padron_costa_resumen = padron_costa[columnas]

    # Renombrar las columnas a snake_case
    padron_costa_resumen.columns = padron_costa_resumen.columns.str.lower().str.replace(' ', '_')

    return padron_costa_resumen
````

````python
# Ejecutar la función
# Ojo: el archivo debe estar en la ubicación ..data/raw..
df_agosto= procesar_padron_ruc('agosto', '2024')
# Verificamos
df_agosto.head()
````

````python
def pct75(column):
    return column.quantile(0.75)

def pct50(column):
    return column.quantile(0.50)

def pct25(column):
    return column.quantile(0.25)

def moda(column):
    moda = column.mode() # función de pandas
    return moda
````

````python
trabajadores = RUC_julio_agosto.pivot_table(values='nrotrab', index='departamento', aggfunc=[pct25, pct50 , pct75, moda])
trabajadores
````
